```python
import os
import schedule
import shutil
import time

# Get the root_path for this jupyter notebook repo.
repo_path = os.path.dirname(os.path.abspath(os.getcwd()))
result_output_folder = os.path.join(repo_path, 'results')

def get_checkpoint_dirs(result_output_folder):
    checkpoints = {}
    for root, dirs, files in os.walk(result_output_folder):
        for dir in dirs:
            if not dir.startswith('checkpoint-'):
                continue
            check_num = int(float(str(dir).split('-')[-1]))
            checkpoints[check_num] = os.path.join(root, dir)
    print('Existing checkpoints include: ')
    print(checkpoints)
    return checkpoints

def get_old_checkpoint_dirs(result_output_folder):
    checkpoints = get_checkpoint_dirs(result_output_folder)
    if not checkpoints:
        return None
    keys = list(checkpoints.keys())
    max_key = max(keys) 
    print(f'Max checkpoint: {max_key}')
    if len(keys) < 2:
        return None
    old_checkpoints = [checkpoints[key] for key in keys if key < max_key]
    return old_checkpoints


def job():
    print("-"*50)
    print("I'm working to remove old checkpoints...")
    repo_path = os.path.dirname(os.path.abspath(os.getcwd()))
    result_output_folder = os.path.join(repo_path, 'results')
    old_checkpoints = get_old_checkpoint_dirs(result_output_folder)
    if not old_checkpoints:
        print("No old checkpoints to remove. We're good, for now...")
        return None
    print(f"Remove {len(old_checkpoints)} old checkpoints")
    for old_checkpoint in old_checkpoints:
        shutil.rmtree(old_checkpoint)
        print(f'Removed: {old_checkpoint}')
    

schedule.every(10).minutes.do(job)

while 1:
    schedule.run_pending()
    time.sleep(1)
    
```
